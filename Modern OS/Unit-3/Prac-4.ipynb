Optimal Page Replacement Technique

```python
def optimal_page_replacement(pages, capacity):
    frames = []
    page_faults = 0

    print("Page\tFrames")

    for i in range(len(pages)):
        page = pages[i]

        # Page Hit
        if page in frames:
            print(f"{page}\t{frames}")
            continue

        page_faults += 1

        # Empty frame available
        if len(frames) < capacity:
            frames.append(page)
        else:
            future_use = {}

            for f in frames:
                if f in pages[i+1:]:
                    future_use[f] = pages[i+1:].index(f)
                else:
                    future_use[f] = float('inf')

            # Replace page with farthest future use
            victim = max(future_use, key=future_use.get)
            frames.remove(victim)
            frames.append(page)

        print(f"{page}\t{frames}")

    print("\nTotal Page Faults:", page_faults)


# Example
pages = [7, 0, 1, 2, 0, 3, 0, 4, 2, 3]
capacity = 3
optimal_page_replacement(pages, capacity)
```
