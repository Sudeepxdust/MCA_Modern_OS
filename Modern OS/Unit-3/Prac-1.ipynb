```python
def fifo_page_replacement(pages, capacity):
    frames = []
    page_faults = 0

    print("Page\tFrames")

    for page in pages:
        if page not in frames:
            page_faults += 1
            if len(frames) < capacity:
                frames.append(page)
            else:
                frames.pop(0)      # Remove oldest page
                frames.append(page)
        print(f"{page}\t{frames}")

    print("\nTotal Page Faults:", page_faults)


# Example
pages = [7, 0, 1, 2, 0, 3, 0, 4]
capacity = 3
fifo_page_replacement(pages, capacity)
```

```python
def lru_page_replacement(pages, capacity):
    frames = []
    page_faults = 0

    print("Page\tFrames")

    for page in pages:
        if page not in frames:
            page_faults += 1
            if len(frames) < capacity:
                frames.append(page)
            else:
                frames.pop(0)      # Remove LRU page
                frames.append(page)
        else:
            frames.remove(page)
            frames.append(page)   # Move to most recently used

        print(f"{page}\t{frames}")

    print("\nTotal Page Faults:", page_faults)


# Example
pages = [7, 0, 1, 2, 0, 3, 0, 4]
capacity = 3
lru_page_replacement(pages, capacity)
```
