```python
import psutil
import time

print("Monitoring Network Statistics... (Press Ctrl+C to stop)\n")

while True:
    net = psutil.net_io_counters()

    print(f"Bytes Sent: {net.bytes_sent}")  
    print(f"Bytes Received: {net.bytes_recv}")
    print("-"*40)

    time.sleep(1)  # Pause for 1 second before the next update
```
