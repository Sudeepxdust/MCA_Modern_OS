```python
import psutil

# Function to convert bytes to GB
def to_gb(value):
    return value / (1024 ** 3)

# Get all disk partitions
partitions = psutil.disk_partitions()

print("Disk Partitions and Usage (in GB):\n")

for partition in partitions:
    print("Partition Device :", partition.device)
    print("Mount Point      :", partition.mountpoint)
    print("File System Type :", partition.fstype)

    # Get usage details of the partition
    usage = psutil.disk_usage(partition.mountpoint)

    print("  Total Size     :", round(to_gb(usage.total), 2), "GB")
    print("  Used           :", round(to_gb(usage.used), 2), "GB")
    print("  Free           :", round(to_gb(usage.free), 2), "GB")
    print("  Usage (%)      :", usage.percent, "%")
    print("-" * 40)
```
