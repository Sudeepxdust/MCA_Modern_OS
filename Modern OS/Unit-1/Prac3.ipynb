```python
import psutil

# Function to convert bytes to GB
def to_gb(value):
    return value / (1024 ** 3)

# Getting virtual memory details
memory_info = psutil.virtual_memory()

print("Total Memory     :", round(to_gb(memory_info.total), 2), "GB")
print("Used Memory      :", round(to_gb(memory_info.used), 2), "GB")
print("Available Memory :", round(to_gb(memory_info.available), 2), "GB")
print("Memory Used (%)  :", memory_info.percent, "%")
```
