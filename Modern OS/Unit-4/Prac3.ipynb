```python
from collections import deque

# Container class
class Container:
    def __init__(self, name, burst):
        self.name = name
        self.burst = burst


# Round Robin Container Scheduler
def container_scheduler(containers, quantum):
    queue = deque(containers)
    time = 0

    print("\n--- Container Scheduling (Round Robin) ---")

    while queue:
        c = queue.popleft()

        if c.burst > quantum:
            time += quantum
            c.burst -= quantum
            print(c.name, "runs for", quantum, "units")
            queue.append(c)
        else:
            time += c.burst
            print(c.name, "runs for", c.burst, "units and finishes")
            c.burst = 0

    print("\nTotal Time =", time)


# ---------------- DRIVER CODE ----------------
containers = [
    Container("Container1", 10),
    Container("Container2", 5),
    Container("Container3", 8)
]

time_quantum = 3

container_scheduler(containers, time_quantum)
```
