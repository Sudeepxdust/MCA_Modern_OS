```python
# Banker's Algorithm in Python

# Allocation Matrix
alloc = [
    [1, 2],
    [2, 0],
    [3, 1]
]

# Maximum Matrix
maxm = [
    [3, 3],
    [3, 2],
    [4, 3]
]

# Available Resources
avail = [2, 1]

# Number of processes and resource types
n = len(alloc)
m = len(avail)

# Calculate Need Matrix
need = [[maxm[i][j] - alloc[i][j] for j in range(m)] for i in range(n)]

finish = [False] * n
safe_seq = []

# Safety Algorithm
while len(safe_seq) < n:
    allocated_in_this_round = False

    for i in range(n):
        if not finish[i] and all(need[i][j] <= avail[j] for j in range(m)):
            # Allocate resources
            for j in range(m):
                avail[j] += alloc[i][j]

            finish[i] = True
            safe_seq.append(i)
            allocated_in_this_round = True

    if not allocated_in_this_round:
        break

# Result
if len(safe_seq) == n:
    print("Safe Sequence:", safe_seq)
else:
    print("System is in UNSAFE state")
```
