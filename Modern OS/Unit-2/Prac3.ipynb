```python
# Priority Scheduling (Non-Preemptive)

processes = []
burst_time = []
priority = []

n = int(input("Enter number of processes: "))

for i in range(n):
    p = "P" + str(i+1)
    processes.append(p)
    bt = int(input(f"Enter Burst Time for {p}: "))
    burst_time.append(bt)
    pr = int(input(f"Enter Priority for {p} (Lower number = Higher priority): "))
    priority.append(pr)

# Combine all data in a list of tuples
combined = list(zip(processes, burst_time, priority))

# Sort by priority (ascending order => highest priority first)
combined.sort(key=lambda x: x[2])

# Unpack sorted data
processes = [x[0] for x in combined]
burst_time = [x[1] for x in combined]
priority = [x[2] for x in combined]

waiting_time = [0] * n
turnaround_time = [0] * n

# Calculate waiting time
for i in range(1, n):
    waiting_time[i] = waiting_time[i-1] + burst_time[i-1]

# Calculate turnaround time
for i in range(n):
    turnaround_time[i] = waiting_time[i] + burst_time[i]

print("\nPriority Scheduling Result:\n")
print("Process\tPriority\tBurst Time\tWaiting Time\tTurnaround Time")

for i in range(n):
    print(f"{processes[i]}\t{priority[i]}\t\t{burst_time[i]}\t\t{waiting_time[i]}\t\t{turnaround_time[i]}")

avg_wait = sum(waiting_time) / n
avg_tat = sum(turnaround_time) / n

print("\nAverage Waiting Time :", avg_wait)
print("Average Turnaround Time :", avg_tat)
```
