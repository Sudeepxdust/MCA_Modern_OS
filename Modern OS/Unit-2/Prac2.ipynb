```python
# Round Robin Scheduling Algorithm

processes = []
burst_time = []

n = int(input("Enter number of processes: "))

for i in range(n):
    p = "P" + str(i+1)
    processes.append(p)
    bt = int(input(f"Enter Burst Time for {p}: "))
    burst_time.append(bt)

time_quantum = int(input("Enter Time Quantum: "))

remaining_bt = burst_time.copy()
waiting_time = [0] * n
turnaround_time = [0] * n

t = 0   # Current time

# Round Robin Execution
while True:
    all_done = True

    for i in range(n):
        if remaining_bt[i] > 0:
            all_done = False

            if remaining_bt[i] > time_quantum:
                t += time_quantum
                remaining_bt[i] -= time_quantum
            else:
                t += remaining_bt[i]
                waiting_time[i] = t - burst_time[i]
                remaining_bt[i] = 0

    if all_done:
        break

# Turnaround time
for i in range(n):
    turnaround_time[i] = burst_time[i] + waiting_time[i]

print("\nRound Robin Scheduling Result:\n")
print("Process\tBurst Time\tWaiting Time\tTurnaround Time")

for i in range(n):
    print(f"{processes[i]}\t{burst_time[i]}\t\t{waiting_time[i]}\t\t{turnaround_time[i]}")

avg_wait = sum(waiting_time) / n
avg_tat = sum(turnaround_time) / n

print("\nAverage Waiting Time :", avg_wait)
print("Average Turnaround Time :", avg_tat)
```
