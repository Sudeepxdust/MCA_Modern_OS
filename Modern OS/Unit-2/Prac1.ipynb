```python
# FCFS Scheduling Algorithm

num = int(input("Enter number of processes: "))

processes = []
burst_time = []

for i in range(num):
    p = "P" + str(i+1)
    processes.append(p)
    bt = int(input(f"Enter Burst Time for {p}: "))
    burst_time.append(bt)

waiting_time = [0] * num
turnaround_time = [0] * num

# Waiting time for FCFS
for i in range(1, num):
    waiting_time[i] = waiting_time[i-1] + burst_time[i-1]

# Turnaround time = waiting + burst
for i in range(num):
    turnaround_time[i] = waiting_time[i] + burst_time[i]

print("\nFCFS Scheduling Result:\n")
print("Process\tBurst Time\tWaiting Time\tTurnaround Time")

for i in range(num):
    print(f"{processes[i]}\t{burst_time[i]}\t\t{waiting_time[i]}\t\t{turnaround_time[i]}")

# Average times
avg_wait = sum(waiting_time) / num
avg_tat = sum(turnaround_time) / num

print("\nAverage Waiting Time :", avg_wait)
print("Average Turnaround Time :", avg_tat)
```
